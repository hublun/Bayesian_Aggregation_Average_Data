```r
# provide your solution here
library("RSQLite")
```

```r
conn <- dbConnect(RSQLite::SQLite(),"CapstoneDB.sqlite")
```

```r
tables <- dbGetQuery(conn, "SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%'")$name

# Drop all tables
for (table in tables) {
  dbExecute(conn, paste0("DROP TABLE IF EXISTS ", table))
}
```

```r
dbExecute(conn, "CREATE TABLE WORLD_CITIES (                            
                            CITY VARCHAR(15) NOT NULL,
                            CITY_ASCII VARCHAR(15) NOT NULL,
                            LAT DECIMAL,
                            LNG DECIMAL,
                            COUNTRY VARCHAR(15),
                            ISO2 CHAR(2),
                            ISO3 CHAR(3),
                            ADMIN_NAME VARCHAR(15),
                            CAPITAL VARCHAR(10),
                            POPULATION INTEGER,
                            CITY_ID CHAR(10) NOT NULL,
                            PRIMARY KEY (CITY_ID))", 
                errors=FALSE)
```

```r
dbExecute(conn, "CREATE TABLE BIKE_SHARING_SYSTEMS (                            
                            COUNTRY VARCHAR(15) NOT NULL,
                            CITY VARCHAR(15) NOT NULL,
                            SYSTEM VARCHAR(15),
                            BICYCLES INTEGER,
                            PRIMARY KEY (COUNTRY, CITY))", 
                errors=FALSE)
```

```r
dbExecute(conn, "CREATE TABLE CITIES_WEATHER_FORECAST (                            
                            CITY VARCHAR(15) NOT NULL,
                            WEATHER VARCHAR(10),
                            VISIBILITY INTEGER,
                            TEMPERATURE DECIMAL,
                            TEMP_MIN DECIMAL,
                            TEMP_MAX DECIMAL,
                            PRESSURE INTEGER,
                            HUMIDITY INTEGER,
                            WIND_SPEED DECIMAL,
                            WIND_DEG INTEGER,
                            SEASON VARCHAR(10),
                            DATETIME DATETIME,
                            PRIMARY KEY (CITY, DATETIME))",

                            errors=FALSE
        )
```

```r
dbExecute(conn, "CREATE TABLE SEOUL_BIKE_SHARING (                            
                            DATE DATE NOT NULL,
                            RENTED_BIKE_COUNT INTEGER,
                            HOUR INTEGER NOT NULL,
                            TEMPERATURE DECIMAL,
                            HUMIDITY INTEGER,
                            WIND_SPEED DECIMAL,
                            VISIBILITY INTEGER,
                            DEW_POINT_TEMP DECIMAL,
                            SOLAR_RADIATION DECIMAL,
                            RAINFALL DECIMAL,
                            SNOWFALL DECIMAL,
                            SEASONS VARCHAR(10),
                            HOLIDAY VARCHAR(15),
                            FUNCTIONING_DAY VARCHAR(10),
                            PRIMARY KEY (DATE, HOUR))",

                            errors=FALSE
        )
```
